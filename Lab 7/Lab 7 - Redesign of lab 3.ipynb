Fetching the ticker symbols for every entity listed in NASDAQ or NYSE from Yahoo Finance 

```python
import requests
import json
import urllib

def get_TickerSymbol(companyName):
    url = "http://d.yimg.com/autoc.finance.yahoo.com/autoc?query={}&region=1&lang=en".format(companyName)

    result = requests.get(url).json()
    
    for x in result['ResultSet']['Result']:
                    return x['symbol']
        

arrTicker=[]
comp_name = df.Entity

    
for  name in comp_name:
    arrTicker.append(get_TickerSymbol(name))
    
    
print(arrTicker)
```

Fetching stock price for the selected period 

```python
start = dt.datetime(2004,1,1)
end = dt.datetime(2017,11,4)
panel_data = pdr.DataReader(arrTicker, 'yahoo', start, end)
```

```python
panel_data
```

```python
# The index in this DataFrame is the major index of the panel_data.
finalData = panel_data.ix['Close']

# Getting all weekdays between 01/01/2000 and 12/31/2016
all_weekdays = pd.date_range(start=start, end=end, freq='B')

# How do we align the existing prices in adj_close with our new set of dates?
# All we need to do is reindex close using all_weekdays as the new index
finalData = finalData.reindex(all_weekdays)
finalData.head(10)
```

Using Transpose to change data Structure 

```python
fd2= finalData.T
fd2
```

Aggregating stock prices by year 

```python
fd2=finalData.groupby(pd.TimeGrouper(freq='12M')).mean()
```

```python
fd2 = fd2.T
fd2
```

Creating Column names and storing data 

```python

df2= pd.DataFrame(columns=['Symbol','Date','Price'])
k =0
for i in range(len(fd2)) : 
    ts = fd2.index[i]
    for j in range(len(fd2.iloc[i])): 
        time_frame = pd.to_datetime(str(fd2.columns.values[j]))
        date = time_frame.strftime('%Y-%m-%d')
        df2.loc[k]=( ts,  date , fd2.iloc[i][j])
        k=k+1
        
```

```python
df2.to_csv('/Users/rosenrot/Downloads/stock.csv', index=True, encoding='utf-8')
```
